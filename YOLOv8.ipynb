# **Training the Model**

```python
!pip install torch
```

```python
import os
import torch  # Import torch for checking CUDA availability
from ultralytics import YOLO

# Ensure your working directory is set to where the data.yaml is located
os.chdir('/content/drive/MyDrive/TEA LEAF')

# Load a pretrained YOLOv8 model for object detection
model = YOLO('yolov8s.pt')  # Ensure the correct weights file is used

# Check if CUDA (GPU) is available
if not torch.cuda.is_available():
    print("Warning: CUDA (GPU) is not available. Training will be slow.")
else:
    print("CUDA (GPU) is available. Training will be fast.")
```

```python
%pwd
```

```python
# Train the model
try:
    model.train(data='data.yaml', epochs=100, imgsz=640, save=True)
    print("Training complete.")
except Exception as e:
    print(f"An error occurred during training: {e}")
```
